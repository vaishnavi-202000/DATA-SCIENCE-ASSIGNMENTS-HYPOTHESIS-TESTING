```python
import pandas as pd
from scipy.stats import chi2_contingency
df = pd.read_csv("Costomer+OrderForm.csv")
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.shape
```

```python
# Count the occurrences of each error type for each country
counts = df.apply(pd.Series.value_counts)
```

```python
# Define colors for each country
colors = ['#66B2FF', '#FF9999', '#99FF99', '#FFCC99']

# Plotting the bar chart with specified colors
plt.figure(figsize=(10, 6))
counts.T.plot(kind='bar', color=colors)
plt.title('Error Types by Country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Error Type')
plt.tight_layout()
plt.show()
```

```python
# Flatten the DataFrame to get a single column of error types
all_errors = df.values.flatten()

# Plotting the histogram of error types
plt.figure(figsize=(8, 6))
sns.histplot(all_errors, bins=2, color='skyblue', discrete=True)
plt.title('Histogram of Error Types Across All Countries')
plt.xlabel('Error Type')
plt.ylabel('Frequency')
plt.show()
```

```python
# Extract value counts for each center
Phillippines_value = df['Phillippines'].value_counts()
Indonesia_value = df['Indonesia'].value_counts()
Malta_value = df['Malta'].value_counts()
India_value = df['India'].value_counts()
```

```python
# Construct the contingency table
contingency_table = pd.DataFrame({
    'Phillippines': Phillippines_value,
    'Indonesia': Indonesia_value,
    'Malta': Malta_value,
    'India': India_value
})
```

```python
# Perform Chi-square test for independence
chi2, pvalue, _, _ = chi2_contingency(contingency_table)

# Print the results
print("Chi-square statistic:", chi2)
print("pvalue:", pvalue)
```

```python
pvalue = 0.2771020991233135
alpha = 0.05
if pvalue < alpha:
    print("H0 is rejected and H1 is accepted.")
else:
    print("H1 is rejected and H0 is accepted.")
```
